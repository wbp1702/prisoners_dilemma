```python
import games
import numpy as np
import matplotlib.pyplot as plt
import multiprocessing

cost_benefit_ratios = [0.1, 0.3, 0.5, 0.7, 0.9]
args = [[100, 100, 50, r, 0.1, 0.1] for r in cost_benefit_ratios]

res = games.nplayer_group_game(100, int(2**16), 50, 0.1, 0.1, 0.1)

plt.plot(res[3])
plt.title("Social Dilemma")
plt.xlabel('Generation')
plt.ylabel('Percentage of Cooperators')
plt.show()


# Plot Outcome B-Values
plt.plot(res[4])
plt.title("Social Dilemma")
plt.xlabel('Generation')
plt.ylabel('Number of Tag Groups')
plt.show()

# Plot Outcome M-Values
plt.plot(res[5])
plt.title("Social Dilemma")
plt.xlabel('Generation')
plt.ylabel('Average Reward Per Round')
plt.show()
```

```python
with multiprocessing.Pool() as pool:
    results = pool.starmap(games.nplayer_group_game, args)

    plt.plot(cost_benefit_ratios, [result[3][-1] for result in results])
    plt.title("Social Dilemma")
    plt.xlabel('r')
    plt.ylabel('Percentage of Cooperators')
    plt.show()


    # Plot Outcome B-Values
    plt.plot(cost_benefit_ratios, [result[4][-1] for result in results])
    plt.title("Social Dilemma")
    plt.xlabel('r')
    plt.ylabel('Number of Tag Groups')
    plt.show()

    # Plot Outcome M-Values
    plt.plot(cost_benefit_ratios, [result[5][-1] for result in results])
    plt.title("Social Dilemma")
    plt.xlabel('r')
    plt.ylabel('Average Reward Per Round')
    plt.show()
```
